```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('data.csv')
df.head()
```

```python
dates = df['<DATE>'].unique()
cols = ['<OPEN>', '<HIGH>', '<LOW>', '<CLOSE>']
new_cols = ['Открытие', 'Макс', 'Мин', 'Закрытие']
df = df.rename(columns = {cols[i]: new_cols[i] for i in range(len(cols))})
```

```python
fig = plt.figure(figsize=(16, 8))
plt.title('Мужик надел шляпу, она ему как раз!!!!')
plt.xticks(rotation = 45)

import seaborn as sns

for i, date in enumerate(dates):
    sns.boxplot(df[df['<DATE>']==date][new_cols].rename(columns = {new_cols[i]:new_cols[i]+date for i in range(len(new_cols))}), legend=True)#.plot(kind = 'box')
```
